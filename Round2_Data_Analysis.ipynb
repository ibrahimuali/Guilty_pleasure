```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
file_path = "/Users/ibrahimuali/Documents/GitHub/Guilty_pleasure/prices_round_2.xlsx"
data_day_minus_1 = pd.read_excel(file_path, sheet_name='prices_round_2_day_-1')
data_day_0 = pd.read_excel(file_path, sheet_name='prices_round_2_day_0')
data_day_1 = pd.read_excel(file_path, sheet_name='prices_round_2_day_1')

# Combine the data into a single DataFrame
combined_data = pd.concat([data_day_minus_1, data_day_0, data_day_1], ignore_index=True)

# Display the first few rows of the combined dataframe to verify the merge
combined_data.head()
```

```python
print(datetime.datetime.fromtimestamp(500))
```

```python
mean1 = combined_data[combined_data['DAY'] == 1]['SUNLIGHT'].mean()
mean2 = combined_data[combined_data['DAY'] == -1]['SUNLIGHT'].mean()
mean0 = combined_data[combined_data['DAY'] == 0]['SUNLIGHT'].mean()

print('Sunlight for day-1: {:.2f} \nSunlight for day0: {:.2f} \nSunlight for day1: {:.2f}'.format(mean2, mean0, mean1))
```

```python
# Create subplots for each day
fig, axs = plt.subplots(3, 1, figsize=(10, 18))

# Plot for DAY -1
axs[0].plot(data_day_minus_1['timestamp'], data_day_minus_1['ORCHIDS'], color='b', label='Orchid Prices')
axs[0].set_xlabel('Timestamp')
axs[0].set_ylabel('Orchid Prices')
axs[0].set_title('Orchid Prices for Day -1')

axs0_secondary = axs[0].twinx()
axs0_secondary.plot(data_day_minus_1['timestamp'], data_day_minus_1['SUNLIGHT'], color='r', label='Sunlight')
axs0_secondary.set_ylabel('Sunlight', color='r')

# Plot for DAY 0
axs[1].plot(data_day_0['timestamp'], data_day_0['ORCHIDS'], color='b', label='Orchid Prices')
axs[1].set_xlabel('Timestamp')
axs[1].set_ylabel('Orchid Prices')
axs[1].set_title('Orchid Prices for Day 0')

axs0_secondary = axs[1].twinx()
axs0_secondary.plot(data_day_0['timestamp'], data_day_0['SUNLIGHT'], color='r', label='Sunlight')
axs0_secondary.set_ylabel('Sunlight', color='r')

# Plot for DAY 1
axs[2].plot(data_day_1['timestamp'], data_day_1['ORCHIDS'], color='b', label='Orchid Prices')
axs[2].set_xlabel('Timestamp')
axs[2].set_ylabel('Orchid Prices')
axs[2].set_title('Orchid Prices for Day 1')

axs0_secondary = axs[2].twinx()
axs0_secondary.plot(data_day_1['timestamp'], data_day_1['SUNLIGHT'], color='r', label='Sunlight')
axs0_secondary.set_ylabel('Sunlight', color='r')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have the following data already defined:
# data_day_minus_1, data_day_0, data_day_1
# Each dataset should contain 'timestamp', 'ORCHIDS', 'SUNLIGHT', and 'HUMIDITY' columns.

# Create a figure and a 3x2 grid of subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 18))  # Increased width to better accommodate two columns

# Configure plots for each day and each environmental factor
for i, data in enumerate([data_day_minus_1, data_day_0, data_day_1]):
    # Plot orchid prices with sunlight on the first column
    ax1 = axs[i, 0]
    ax1.plot(data['timestamp'], data['ORCHIDS'], color='b', label='Orchid Prices')
    ax1.set_xlabel('Timestamp')
    ax1.set_ylabel('Orchid Prices', color='b')
    ax1.set_title(f'Orchid Prices and Sunlight for Day {i-1}')
    ax1_secondary = ax1.twinx()
    ax1_secondary.plot(data['timestamp'], data['SUNLIGHT'], color='r', label='Sunlight')
    ax1_secondary.set_ylabel('Sunlight', color='r')
    
    # Plot orchid prices with humidity on the second column
    ax2 = axs[i, 1]
    ax2.plot(data['timestamp'], data['ORCHIDS'], color='b', label='Orchid Prices')
    ax2.set_xlabel('Timestamp')
    ax2.set_ylabel('Orchid Prices', color='b')
    ax2.set_title(f'Orchid Prices and Humidity for Day {i-1}')
    ax2_secondary = ax2.twinx()
    ax2_secondary.plot(data['timestamp'], data['HUMIDITY'], color='g', label='Humidity')
    ax2_secondary.set_ylabel('Humidity', color='g')

# Improve the layout to prevent label overlap and add a legend
fig.tight_layout()
plt.savefig("Orchids_P.png")
plt.show()
```
